## **3. Data Wrangling**

```python
# ============================================================
# DATA WRANGLING
# Goal: Make dataset clean and analysis-ready
# ============================================================

# Step 1: Select only relevant columns for EDA
cols_needed = [
    'iyear', 'imonth', 'iday', 'country_txt', 'region_txt', 'city',
    'latitude', 'longitude', 'attacktype1_txt', 'targtype1_txt',
    'gname', 'weaptype1_txt', 'nkill', 'nwound', 'success', 'ishostkid', 'property'
]
df_eda = df[cols_needed].copy()
print(f"Step 1 ✅ Selected {len(cols_needed)} key columns.")

# Step 2: Fill missing numerical values (kills/wounds) with 0
# Rationale: Missing values here likely mean 0 casualties were reported
df_eda['nkill'] = df_eda['nkill'].fillna(0)
df_eda['nwound'] = df_eda['nwound'].fillna(0)
print("Step 2 ✅ Filled missing nkill and nwound with 0.")

# Step 3: Create a 'total_casualties' derived feature
df_eda['total_casualties'] = df_eda['nkill'] + df_eda['nwound']
print("Step 3 ✅ Created 'total_casualties' column (nkill + nwound).")

# Step 4: Create a 'decade' feature for trend analysis
df_eda['decade'] = (df_eda['iyear'] // 10) * 10
df_eda['decade'] = df_eda['decade'].astype(str) + 's'
print("Step 4 ✅ Created 'decade' column.")

# Step 5: Replace missing ishostkid with 0 (no hostage situation)
df_eda['ishostkid'] = df_eda['ishostkid'].fillna(0)
print("Step 5 ✅ Filled missing 'ishostkid' with 0.")

# Step 6: Drop rows where region or attacktype is missing (critical fields)
before = len(df_eda)
df_eda.dropna(subset=['region_txt', 'attacktype1_txt'], inplace=True)
after = len(df_eda)
print(f"Step 6 ✅ Dropped {before - after} rows with missing region/attacktype.")

print(f"\n🔷 Final dataset shape after wrangling: {df_eda.shape}")
```

### **What manipulations have you done and insights found?**

- **Column Selection**: Reduced the 135-column dataset to 17 key columns most relevant for EDA, making analysis faster and cleaner.
- **Missing Value Imputation**: `nkill` and `nwound` NaN values were filled with 0 since missing casualties logically imply no reported deaths/injuries.
- **Feature Engineering**:
  - `total_casualties` = `nkill` + `nwound` — gives a unified impact measure per attack.
  - `decade` — groups years into decades for trend analysis (1970s, 1980s, etc.).
- **Row Dropping**: Only rows missing critical categorical fields (`region_txt`, `attacktype1_txt`) were removed — a negligible fraction of the data.
- **Key Insight**: After wrangling, the dataset is clean, consistently typed, and enriched with derived features ready for visualization.

### 📊 Chart 1 — Distribution of Attacks Per Year (Univariate - Numerical)

```python
# Chart 1: Number of Terrorist Attacks Per Year
attacks_per_year = df_eda.groupby('iyear').size().reset_index(name='attack_count')

plt.figure(figsize=(16, 5))
plt.plot(attacks_per_year['iyear'], attacks_per_year['attack_count'],
         color='crimson', linewidth=2, marker='o', markersize=3)
plt.fill_between(attacks_per_year['iyear'], attacks_per_year['attack_count'],
                 alpha=0.2, color='crimson')
plt.title('Number of Terrorist Attacks Per Year (1970–2017)', fontsize=15, fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Number of Attacks')
plt.xticks(range(1970, 2018, 5), rotation=45)
plt.tight_layout()
plt.show()
```

**1. Why did you pick this chart?**  
A line chart is ideal for showing a continuous time series — it clearly reveals trends, peaks, and dips in attack frequency over the 47-year span.

**2. Insights found:**  
- There is a sharp escalation in attacks post-2011, with a peak around 2014–2015, coinciding with the rise of ISIS.
- A significant drop occurred in the 1990s after the Cold War ended.
- The 1970s and early 1980s had moderate activity primarily from left-wing and separatist groups.
- Negative insight: The dramatic spike post-2010 indicates terrorism became more widespread and frequent, suggesting prior counter-terrorism efforts were insufficient.

**3. Business Impact:**  
This trend clearly shows that resources for counter-terrorism need to be dramatically scaled post-2010. Intelligence agencies can use this trend to forecast future attack frequency and budget accordingly.

### 📊 Chart 10 — Kills vs Wounds Scatter Plot (Bivariate - Numerical vs Numerical)

```python
import numpy as np
# Chart 10: nkill vs nwound scatter

# Filter data to include only attacks with both kills and wounds > 0
filtered_data = df_eda[(df_eda['nkill'] > 0) & (df_eda['nwound'] > 0)]

# Determine sample size: take the minimum of 5000 and the actual number of filtered rows
sample_size = min(5000, len(filtered_data))

# Take a sample if the filtered data is not empty
if sample_size > 0:
    sample = filtered_data.sample(sample_size, random_state=42)
    plt.figure(figsize=(10, 6))
    sns.scatterplot(data=sample, x='nkill', y='nwound',
                    alpha=0.4, color='steelblue', edgecolor='none')
    plt.title('Kills vs Wounds per Attack', fontsize=14, fontweight='bold')
    plt.xlabel('Number Killed')
    plt.ylabel('Number Wounded')
    plt.xscale('log')
    plt.yscale('log')
    plt.tight_layout()
    plt.show()
else:
    print("No data points with both kills and wounds > 0 to plot.")
```

**1. Why did you pick this chart?**  
A scatter plot is the standard way to visualize the relationship between two continuous numerical variables. Log scales prevent overplotting at low values.

**2. Insights found:**  
- There is a positive correlation between kills and wounds — attacks that kill more also tend to wound more.
- Some attacks have very high wounds but low kills, suggesting non-lethal but injurious explosions.
- Negative insight: High-kill, high-wound events represent mass casualty incidents that strain medical infrastructure.

**3. Business Impact:**  
Medical emergency planning should account for both kills AND wounds simultaneously — a single large bombing can overwhelm hospital capacity across both metrics.

## ML IMPLEMENTATION

Logistic Regression

Theory:
Logistic Regression is used when the output is yes/no or 0/1.
It predicts the probability of an event happening using a special curve called the sigmoid function.

Example (GTD):
Predict whether an attack will be successful or not.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

df = pd.read_csv("globalterrorismdb_0718dist.csv", encoding='latin-1').fillna(0)

# Use 2 features for graph
X = df[['iyear', 'imonth']]
y = df['success']

model = LogisticRegression()
model.fit(X, y)

plt.figure()
plt.scatter(X['iyear'], X['imonth'])

# Decision boundary
x_vals = [X['iyear'].min(), X['iyear'].max()]
y_vals = -(model.coef_[0][0]*pd.Series(x_vals) + model.intercept_) / model.coef_[0][1]

plt.plot(x_vals, y_vals)

plt.xlabel("Year")
plt.ylabel("Month")
plt.title("Logistic Regression Decision Boundary")
plt.show()
```

Decision Tree

Theory:
Decision Tree works like a tree structure or flowchart.
It asks questions step by step (like yes/no) and divides data into smaller parts to make a decision.

Example (GTD):
Classify attacks based on type, location, or year.

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree

model = DecisionTreeClassifier(max_depth=3)
model.fit(X, y)

plt.figure(figsize=(10,6))
plot_tree(model, filled=True)
plt.title("Decision Tree")
plt.show()
```

Random Forest

Theory:
Random Forest is a group of many decision trees.
Each tree gives its answer, and the final answer is based on majority voting.
It is more accurate and reduces mistakes compared to a single tree.

Example (GTD):
Better prediction of attack success or category.

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(X, y)

importance = model.feature_importances_

plt.figure()
plt.bar(['iyear', 'imonth'], importance)
plt.xlabel("Features")
plt.ylabel("Importance")
plt.title("Feature Importance (Random Forest)")
plt.show()
```

K-Means Clustering

Theory:
K-Means is used when there are no labels (unsupervised learning).
It groups similar data points into clusters based on similarity.

Example (GTD):
Group similar types of attacks or regions.

```python
from sklearn.cluster import KMeans

X = df[['iyear', 'nkill']].fillna(0)

kmeans = KMeans(n_clusters=3)
df['cluster'] = kmeans.fit_predict(X)

plt.figure()
plt.scatter(X['iyear'], X['nkill'])
plt.xlabel("Year")
plt.ylabel("Casualties")
plt.title("K-Means Clusters")
plt.show()
```

Linear Regression
Theory:
Linear Regression is used to predict numerical values.
It finds a straight line that shows the relationship between input and output.
Example (GTD):
Predict number of casualties in an attack.




```python
from sklearn.linear_model import LinearRegression

X = df[['iyear']]
y = df['nkill'].fillna(0)

model = LinearRegression()
model.fit(X, y)

plt.figure()
plt.scatter(X, y)

# Line
plt.plot(X, model.predict(X))

plt.xlabel("Year")
plt.ylabel("Casualties")
plt.title("Linear Regression Line")
plt.show()
```
